```python
import pandas as pd
df = pd.read_excel('Lasso_params_transp.xlsx', index_col=0)
df
```

```python
adm = ['Гуманитарными науками (филология, журналистика, культурология и т.д.)', 'Информационными технологиями', 
       'Экономикой', 'Педагогикой', 'Медициной', 'Естественными науками (физика, химия и т. д.)', 'Инженерией']
subjects = ['Алгебра', "Геометрия", "Русский язык", "Литература", "Физика", "Химия", "Биология", "География", "Иностранный язык", "Информатика", "Обществознание", "История", "Ничего"]
olymps = ["Математике", "Русский языку", "Литературе", "Физике", "Химии", "Биологии", "Географии", "Иностранный языку", "Информатике", "Обществознанию", "Истории", "Нигде"]
adm_short = ['gum', 'it', 'econ', 'ped', 'med', 'est', 'ing']
subj_short = ['alg', 'geom', 'rus', 'lit', 'phy', 'chem', 'bio', 'geo', 'for', 'it', 'soc', 'his', 'no']
olymps_short = ['math', 'rus', 'lit', 'phy', 'chem', 'bio', 'geo', 'for', 'it', 'soc', 'his', 'no']
vol = ['Волонтерском движении', 'Спортивных мероприятиях', 'Общественных мероприятиях']
vol_short = ['vol', 'sport', 'events']
subjects_eng = 'algebra	geometry	russian	literature	history	foreign	physics	chemystry	geography	social	biology	informatics'.split('	')
```

```python
df_rus = pd.DataFrame(columns=['Положительная корреляция','Отрицательная корреляция'])
for adm_1 in list(df.columns.values):
    short_df = df[df[adm_1]!=0][adm_1]
    adm_rus=adm[adm_short.index(adm_1)]
    pluses = []
    minuses = []
    alls = short_df.index
    for one in alls:
        one_rus = ''
        base_adv_ol = ''
        if len(one.split('_'))==2:
            if one.split('_')[0]!='math':
                ssss=one.split('_')[1]
                if ssss=='adv':
                    base_adv_ol = "продвинутый уровень"
                elif ssss=='base':
                    base_adv_ol = 'базовый уровень'
                elif ssss=='town':
                    base_adv_ol = 'участие в городском этапе олимпиады'
                elif ssss=='region':
                    base_adv_ol = 'участие в региональном этапе олимпиады'
                elif ssss=='state':
                    base_adv_ol = 'участие в финальном этапе олимпиады'
                elif ssss=='other':
                    base_adv_ol = 'участие в олимпиадах вузах'
                one_rus=subjects[subj_short.index(one.split('_')[0])] + ', ' + base_adv_ol
            elif one.split('_')[0]=='math':
                ssss=one.split('_')[1]
                if ssss=='town':
                    base_adv_ol = 'участие в городском этапе олимпиады'
                elif ssss=='region':
                    base_adv_ol = 'участие в региональном этапе олимпиады'
                elif ssss=='state':
                    base_adv_ol = 'участие в финальном этапе олимпиады'
                elif ssss=='other':
                    base_adv_ol = 'участие в олимпиадах в вузах'
                one_rus="Математика" + ', ' + base_adv_ol
        elif len(one.split('_'))==1:
            one_rus=subjects[subjects_eng.index(one.split('_')[0])] + ', оценка'  
        if short_df[one] > 0:
            pluses.append(one_rus)
        else:
            minuses.append(one_rus)
    print('Предмет:'+ adm_rus)
    print('\n'.join(pluses[:5]))
    print('\n')
    print('\n'.join(minuses[-5:]))
    print('\n')
    df_rus = df_rus.append(pd.Series({'Положительная корреляция': '\n'.join(pluses[:5]),
                                      'Отрицательная корреляция': '\n'.join(minuses[-5:])}, 
                                     name = adm_rus))
```

```python
df_rus.to_excel('correlations.xlsx')
```
